```python
# Task 2

import numpy as np
import matplotlib.pyplot as plt
GAMMA = 1.4
Cp = 1005   # J/(kg*K)
R = 287
M = np.arange(0.1,10,0.01)
M2 = np.arange(0,3,0.1)
STAGNATION_TEMP_TO_SONIC_STAGNATION_TEMP = ((GAMMA+1)*(M**2)*(2+(GAMMA-1)*M**2)/((1+(GAMMA*M**2))**2))
STATIC_TEMP_TO_SONIC_TEMP = (((1+GAMMA)**2)*(M**2))/(1+(GAMMA*M**2))**2
STATIC_PRESSURE_TO_SONIC_STATIC_PRESSURE = (1+GAMMA)/(1+(GAMMA*M**2))
SONIC_DENSITY_TO_STATIC_DENSITY = ((1+GAMMA)*M**2)/(1+(GAMMA*M**2))
STAGNATION_PRESSURE_TO_SONIC_STAGNATION_PRESSURE = (((2/(GAMMA+1))*(1+((GAMMA-1)/2)*M2**2))**(GAMMA/(GAMMA-1)))*((1+GAMMA)/(1+(GAMMA*M2**2)))


plt.plot(M,STAGNATION_TEMP_TO_SONIC_STAGNATION_TEMP, color='red',label='T0/T0*')
plt.plot(M,STATIC_TEMP_TO_SONIC_TEMP,color='purple',label='T/T*')
plt.plot(M,STATIC_PRESSURE_TO_SONIC_STATIC_PRESSURE,color='green',label = 'P/P*')
plt.plot(M,SONIC_DENSITY_TO_STATIC_DENSITY,color='black',label='rho*/rho')
plt.plot(M2,STAGNATION_PRESSURE_TO_SONIC_STAGNATION_PRESSURE,color='magenta',label='P0/P0*')
plt.xlabel('MACH NUMBER')
plt.ylabel('RATIOS')
plt.title(' DIABATIC PROPERTY RATIO vs MACH NUMBER ')
plt.legend()
plt.show()

STATIC_TEMP_TO_SONIC_TEMP = (((1+GAMMA)**2)*(M**2))/(1+(GAMMA*M**2))**2
STATIC_PRESSURE_TO_SONIC_STATIC_PRESSURE = (1+GAMMA)/(1+(GAMMA*M**2))
DIMENTIONLESS_ENTROPY =(np.log(STATIC_TEMP_TO_SONIC_TEMP))-((GAMMA-1)/GAMMA)*(np.log(STATIC_PRESSURE_TO_SONIC_STATIC_PRESSURE))
plt.figure(figsize=(10, 6))
plt.plot(DIMENTIONLESS_ENTROPY,STATIC_TEMP_TO_SONIC_TEMP, color='GREEN',label='T/T* vs (S-S*)/Cp')
plt.xlabel('(S-S*)/Cp')
plt.ylabel('T/T*')
plt.title('Temperature Vs Entropy')
plt.show()
```

```python
# TASK 3

import numpy as np
GAMMA = 1.4
Cp = 1005   # J/(kg*K)
R = 287
M1 = float(input("Enter Upstream Mach Number: "))
P1 = float(input("Enter Upstream Pressure: " )) # Pa
T1 = float(input("Enter Upstream Temperature: "))   # K
q = float(input("Enter Amount of Heat added per unit  mass: ")) # J/kg

T01 = T1 * (1 + ((GAMMA-1)/2)*(M1)**2)
P01 = P1*((T01/T1)**((GAMMA)/(GAMMA-1)))
T02 = (q/Cp) + T01
# T1_Tsonic = (((1+GAMMA)**2)*(M1**2))/(1+(GAMMA*M1**2))**2
Tsonic = T1/((((1+GAMMA)**2)*(M1**2))/(1+(GAMMA*M1**2))**2)
# P1_Psonic = (1+GAMMA)/(1+(GAMMA*M1**2))
Psonic = P1/((1+GAMMA)/(1+(GAMMA*M1**2)))
# T01_T0sonic = ((GAMMA+1)*(M1**2)*(2+(GAMMA-1)*M1**2)/((1+(GAMMA*M1**2))**2))
T0sonic = T01/(((GAMMA+1)*(M1**2)*(2+(GAMMA-1)*M1**2)/((1+(GAMMA*M1**2))**2)))
# P01_P0sonic = (((2/(GAMMA+1))*(1+((GAMMA-1)/2)*M1**2))**(GAMMA/(GAMMA-1)))*((1+GAMMA)/(1+(GAMMA*M1**2)))
P0sonic = P01/((((2/(GAMMA+1))*(1+((GAMMA-1)/2)*M1**2))**(GAMMA/(GAMMA-1)))*((1+GAMMA)/(1+(GAMMA*M1**2))))

X = T02/T0sonic
Prod1 = (((2*GAMMA*(1-X))+2)/((GAMMA*GAMMA*(X-1))+1))
Prod2 = ((((((2*GAMMA*(1-X))+2)/((GAMMA*GAMMA*(X-1))+1))**2)-((4*X)/((GAMMA*GAMMA*(X-1))+1)))**0.5)
if M1< 1:
    M2 = ((Prod1-Prod2)/2)**0.5
else:
    M2 = ((Prod1+Prod2)/2)**0.5

T2_T0sonic = (((1+GAMMA)**2)*(M2**2))/(1+(GAMMA*M2**2))**2
P2_P0sonic = (1+GAMMA)/(1+(GAMMA*M2**2))
P02_P0sonic = (((2/(GAMMA+1))*(1+((GAMMA-1)/2)*M2**2))**(GAMMA/(GAMMA-1)))*((1+GAMMA)/(1+(GAMMA*M2**2)))

T2 = Tsonic*T2_T0sonic
P2 = Psonic*P2_P0sonic
P02 = P0sonic*P02_P0sonic
RHO2 = P2/(R*T2)

print("Stagnation Temperature at station 1:",T01)
print("Stagnation Temperature at station 1:",T02)
print("Mach Number at station 2:",M2)
print("Static Pressure at station 2:",P2)
print("Static Temperature at station 2:",T2)
print("Density at station 2:",RHO2)
print("Stagnation Temperature at station 2:",T02)
print("Stagnation Pressure at station 2:",P02)
```
